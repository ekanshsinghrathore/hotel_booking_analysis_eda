```python
#df=pd.read_csv('/content/drive/MyDrive/Hotel Bookings.csv')
url = 'https://drive.google.com/file/d/1C9AxF9fcVzMw0Bgs0NaRrNML2WwX1Ehm/view?usp=sharing'
file_id=url.split('/')[-2]
dwn_url='https://drive.google.com/uc?id=' + file_id
df = pd.read_csv(dwn_url)
```

# 1.As we can see from the above graph, people are very interested in staying at city hotels instead of resort hotels, and there are approximately 50% fewer bookings at resort hotels as compared to city hotels.

```python
sns.countplot(x='hotel',data=df)
plt.show()
```

# 2. The graph below shows us which months are the most popular amongÂ guests visiting hotels.The below graph shows the highest bookings recorded in the month of August and the third quarter of the year, which was the busiest time of the year. From the following graph, we can conclude that the third quarter is the best time to visit the place, and March is the best month for visiting because the rush has decreased significantly.

```python
def numbermonths(string):
    month_dict={}
    for x in string:
      if x in month_dict.keys():
        month_dict[x]=month_dict[x]+1
      else:
        month_dict[x]=1
    return month_dict
```

```python
arrival_date_month_dict=numbermonths(df['arrival_date_month'])
```

```python
arrival_date_month_dict_sort={'January': 5929,'February': 8068,'March': 9794,'April': 11089,'May': 11791,'June': 10939,'July': 12661,'August': 13877,'September': 10508,'October': 11160,'November': 6794,'December': 6780} 
```

```python
plt.xticks(rotation=90)
plt.bar(arrival_date_month_dict_sort.keys(),arrival_date_month_dict_sort.values())
plt.show()
```

# 3.As we can see from the graph below, booking frequency for resort and city hotels follows the same pattern.


```python
sns.set_theme(style="whitegrid")
sns.countplot(x='hotel',hue='arrival_date_month',data=df)
plt.show()
```

# 4.From this boxplot we can easily identify that mean lead time for booking is 100 hrs (approx.).70 to 160 hours lead time is very common in graphs. Mostly guests prefer booking in this lead time range and there are also very little number of people who prefer booking in 1 month in advance.

```python
sns.boxplot(y=df['lead_time'])
plt.show()
```

# 5.As we can see from the count plot given below that majority of the bookings were not cancelled, but still there is little amount of bookings cancelled.

```python
df['is_canceled'] = df.is_canceled.replace([1,0], ['canceled', 'not_canceled'])
booking_canceled_data = df['is_canceled']
sns.countplot(booking_canceled_data)
plt.show()
```
